## Preprocessing For All the files

```python
notes = []
for file in glob.glob("midi_songs/*.mid"):
    midi = converter.parse(file) # Convert File into stream.Score Object which is a container containing files
    print("Parsing %s"%file)
    
    elements_to_parse = midi.flat.notes
    
    for element in elements_to_parse:
        # If the element is a Note, then store its Pitch (like A4, G4)
        if isinstance(element,note.Note):
            notes.append(str(element.pitch))
        # If the element is a chord, split each note of chord and join them with + or any other operater
        elif isinstance(element,chord.Chord):
            notes.append("+".join(str(n) for n in element.normalOrder))
            
```

## Prepare Sequential Data For LSTM

```python
# How Many elemnets LSTM input should consider
sequence_length = 100
```

```python
pitch_names = sorted(set(notes))
```

Any ML/DL doesn't work with strings so we will make integers

```python
## Mapping between elements to int values
ele_to_int = dict((ele,num) for num,ele in enumerate(pitch_names))
int_to_ele = dict((num,ele) for num,ele in enumerate(pitch_names))
```

```python
network_input = []
network_output = []
```

```python
for i in range(len(notes) - sequence_length):
    seq_in = notes[i:i+sequence_length] # contains 100 values
    seq_out = notes[i+sequence_length]
    
    network_input.append([ele_to_int[ch] for ch in seq_in])
    network_output.append(ele_to_int[seq_out])
```

```python
print(len(network_input[0]))
print(len(network_input))
print(len(network_output)) 
```

```python
# Number of examples
n_patterns = len(network_input)
print(n_patterns)
```

```python
network_input = np.reshape(network_input,(n_patterns,sequence_length,1))
print(network_input.shape)
```

```python
network_input  ## We have values from 0-358 , sso we can normaize it
```

```python
## Normalize input (getting values between 0 and 1)
normalize_network_input = network_input/float(n_vocab)
```

```python
normalize_network_input
```

```python
network_output
```

```python
## We have 359 classes in network output so we will convert it into one hot encoded vector
network_output = np_utils.to_categorical(network_output)
print(network_output.shape)
```

```python
print(normalize_network_input.shape)
print(network_output.shape)
```

## Create Model

```python
from keras.models import Sequential
from keras.layers import *
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
model = Sequential()
## Input has shape 100,1
model.add(LSTM(units=512,
               input_shape = (normalize_network_input.shape[1],normalize_network_input.shape[2]),
               return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(512,return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(n_vocab,activation = 'softmax')) # Last layer
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam")
```

```python
model.summary()
```

```python
checkpoint = ModelCheckpoint("model.hdf5",monitor='loss',verbose=0,save_best_only=True,mode='min')
model_hist = model.fit(normalize_network_input,network_output,epochs=100,batch_size=64,callbacks=[checkpoint])
```

- We are going to pick any random data point which will have 100 elements. We will take this as a start and will start generating outpus.

```python
## Initial Pattern
start = np.random.randint(len(network_input)-1)
pattern = network_input[start]
prediction_output = []

## How many elements we generate is a hyperparameter
# Lets Generate 200 elements here
for note_index in range(200):
    prediction_input = np.reshape(pattern, (1,len(pattern),1))
    ##Batch Size 1 is data point()one data point,100 points, last one due to 3d struture as LSTM Accepts in 3D format
    prediction_input = prediction_input/float(n_vocab)
    prediction = model.predict(prediction_input,verbose=0)
    idx = np.argmax(prediction)
    result = int_to_ele[idx]
    prediction_output.append(result)
    # Now change input sequence
    pattern.append(idx)
    pattern = pattern[1:]
```

```python
# offset is the position of a particular note
offset = 0 # Also treated as Time
output_notes = []

for pattern in prediction_output:
    # if pattern is a chord
    if('+' in pattern) or pattern.isdigit():
        notes_in_chord = pattern.split('+')
        temp_notes = []
        for current_note in notes_in_chord:
            new_note = note.Note(int(current_note)) # Create Note object for each note in the chord
            new_note.storedInstrument = instrument.Piano()
            temp_notes.append(new_note)
        
        new_chord = chord.Chord(temp_notes) # Creates the chord obejct from the list of notes
        new_chord.offset = offset
        output_notes.append(new_chord)
        
    # if it is a note
    else:
        new_note = note.Note(pattern)
        new_note.offet = offset
        new_note.storedInstrument = instrument.Piano()
        output_notes.append(new_note)
        
    offset+=0.5 ## Right now we are fixed i.e every note will be played for 0.5 offset only
```
